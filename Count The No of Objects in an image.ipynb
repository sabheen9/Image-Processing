# Count The No of Objects in an image

### Importing Liabraries

```python
from skimage.filters import threshold_otsu
from skimage import data
import matplotlib.pyplot as plt
from skimage import measure
import numpy as np
```

### Define a function to show images

```python
#Define a function to show images

def show_images(image,title):
    plt.imshow(image,cmap='gray')
    plt.axis('off')
    plt.title(title)
    plt.show()
```

```python
image=data.coins()
show_images(image,"coins")
```

```python
# checking the shape of the imgae
image.shape

# becasue image shape dont have rgb layer so it means that this is the gray image so we dont need to transfer ot to gray color
```

### Binarizing the Image

```python
thresh=threshold_otsu(image)
binary_image=image>thresh
show_images(binary_image,"Threshold Image")
show_images(image,"Original Image")
```

### Contouring the Image

```python
contours=measure.find_contours(binary_image,0.8)

# 0.8 is a constant value .The level value varies between 0 and 1 , the closer to 1 the more sensitive the method in detecting contours.
# More complex contours can be detected

for contour in contours:
    print(contour.shape)

print(len(contours))
```

### Counting the objects

```python
# Create list with the shape of each contour
shape_contours = [cnt.shape[0] for cnt in contours]

# Set 150 as the minimum size of the dots shape
max_coin_shape = 150

# Count coins in contours excluding smaller than coin size
coin_contours = [cnt for cnt in contours if (np.shape(cnt)[0] > max_coin_shape) and (np.shape(cnt)[0] < 1039)  ]

# Print the Coin's number
print("coins's dots number: {}. ".format(len(coin_contours)))
```
